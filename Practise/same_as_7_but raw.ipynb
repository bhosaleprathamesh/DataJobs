```python
#Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup 
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.head(5)
```

```python
df_india = df[df['job_country'] == 'India']
df_india.head(6)
```

```python
df_india['job_month'] = df_india.job_posted_date.dt.month_name
df_india.sample(n=6)
```

```python
df_india.loc[:, 'job_month'] = df_india['job_posted_date'].dt.month_name()
```

```python
df_india.sort_index()
df_india.head(6)
```

```python
india_job_count = df_india.pivot_table(index='job_title_short', values='job_posted_date', columns='job_month', aggfunc='count')
india_job_count
```

```python
india_job_count = df_india.pivot_table(index='job_month', values='job_posted_date', columns='job_title_short', aggfunc='count')
india_job_count
```

```python
india_job_count.sort_index()
```

```python
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']

india_job_count = india_job_count[job_titles]
india_job_count
```

```python
india_job_count.plot(kind='line')
```
