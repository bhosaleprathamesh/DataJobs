```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_country'].value_counts()
```

```python
df_da_uk = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United Kingdom')].copy()
df_da_uk
```

```python
df_da_uk = df_da_uk.dropna(subset='salary_year_avg')
df_da_uk
```

```python
df_da_uk = df_da_uk.explode('job_skills')
df_da_uk
```

```python
df_top_pay = df_da_uk.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='median', ascending=False).head(10)
df_top_pay
```

```python
df_top_pay[::-1].plot(kind='barh', y='median')
```

```python
df_skills = df_da_uk.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='count', ascending=False).head(10)
df_skills = df_skills.sort_values(by='median', ascending=False)
df_skills
```

```python
fig, ax = plt.subplots(2, 1)

df_top_pay[::-1].plot(kind='barh', y='median', legend=False, ax=ax[0])
ax[0].set_title('Top 10 Highest Paid Skills For Data Analyst in UK 2023')
ax[0].set_ylabel('')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

df_skills[::-1].plot(kind='barh', y='median', legend=False, ax=ax[1])
ax[1].set_title('Top 10 Most In-Demand Skills For Data Analyst in UK 2023')
ax[1].set_ylabel('')
ax[1].set_xlabel('Median Salary ($USD)')
ax[1].set_xlim(ax[0].get_xlim())
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

plt.tight_layout()
plt.show()
```
