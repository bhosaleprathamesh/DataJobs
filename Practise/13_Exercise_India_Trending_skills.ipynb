```python
#Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup 
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_india = df[df['job_country'] == 'India'].copy()
df_india
```

```python
df_da_india = df_india[df_india['job_title_short'] == 'Data Analyst']
df_da_india
```

```python
df_explode = df_da_india.explode('job_skills')
df_explode
```

```python
df_explode['job_month'] = df_explode['job_posted_date'].dt.to_period('M')
df_explode
```

```python
df_pivot = df_explode.pivot_table(index='job_month', columns='job_skills', aggfunc='size', fill_value=0)
df_pivot
```

```python
df_pivot.loc['Total'] = df_pivot.sum()
df_pivot
```

```python
df_pivot.loc['Total'].sort_values(ascending=False)
```

```python
df_pivot = df_pivot[df_pivot.loc['Total'].sort_values(ascending=False).index]
df_pivot
```

```python
df_pivot = df_pivot.drop('Total')
df_pivot
```

```python
df_pivot.iloc[:, :5].plot(kind='line')
plt.title('Top 5 skills for Data Analyst Per Month in India')
plt.xlabel('')
plt.ylim(0, 500)
plt.show()
```
