```python
#Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup 
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_da = df[df['job_title_short'] == 'Data Analyst']
df_da
```

```python
df_exploded = df_da.explode('job_skills')
df_exploded
```

```python
df_exploded.groupby('job_skills').agg(
    skill_count=('job_skills', 'count'),
    median_salary=('salary_year_avg', 'median')
)
```

```python
skill_stats = df_exploded.groupby('job_skills').agg(
    skill_count=('job_skills', 'count'),
    median_salary=('salary_year_avg', 'median')
)

skill_stats = skill_stats.sort_values(by='skill_count', ascending=False).head(10)
skill_stats
```

```python
skill_stats.plot(kind='scatter', x='skill_count', y='median_salary')
plt.xlabel('Count of Job Postings')
plt.ylabel('Median yearly Salary ($USD)')
plt.title('Salary vs Count of Job Postings for Top 10 Skills')
plt.tight_layout()
plt.show()
```

```python
#now to label each coordinate we need to use for loop
#first to check if for loop is working
for i, txt in enumerate(skill_stats.index):
    print(i, txt)
```

```python
skill_stats.plot(kind='scatter', x='skill_count', y='median_salary')
plt.xlabel('Count of Job Postings')
plt.ylabel('Median yearly Salary ($USD)')
plt.title('Salary vs Count of Job Postings for Top Skills in Data Analyst Jobs')

for i, txt in enumerate(skill_stats.index):
    plt.text(skill_stats['skill_count'].iloc[i], skill_stats['median_salary'].iloc[i], txt)
#plt.text(x, y, 'text')

plt.tight_layout()
plt.show()
```

```python
top_7 = df.job_country.value_counts().head(7).index.to_list()
top_7
```

```python
df_top_7 = df[df['job_country'].isin(top_7)]
df_top_7.job_country.value_counts()
```

```python
df_top_7_exploded = df_top_7.explode('job_skills')
df_top_7_exploded
```

```python
df_top_7_exploded.groupby('job_skills').agg(
    skill_count = ('job_skills', 'count'),
    median_salary = ('salary_year_avg', 'median')
)
```

```python
df_grpby = df_top_7_exploded.groupby('job_skills').agg(
    skill_count = ('job_skills', 'count'),
    median_salary = ('salary_year_avg', 'median')
)

df_grpby
```

```python
df_grpby.sort_values(by='skill_count', ascending=False, inplace=True)
df_grpby = df_grpby.head(10)
df_grpby
```

```python
df_grpby.plot(kind='scatter', x='skill_count', y='median_salary')
```

```python
for i, text in enumerate(df_grpby.index):
    print(i, text)
```
