```python
#Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup 
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df.explode('job_skills')
```

```python
df_exploded = df.explode('job_skills')
df_exploded
```

```python
df_exploded.value_counts('job_skills').head(10).plot(kind='bar')
```

But this bar chart is calculating the skills for all jobs we want to group them not only by their skill count but also their job title short count

```python
df_exploded.groupby(['job_title_short', 'job_skills']).size()
```

```python
skills_count = df_exploded.groupby(['job_title_short', 'job_skills']).size()
type(skills_count)
```

```python
#change the skills_count to a dataframe cuz it is easier to plot from dataframe 
df_skills_count = skills_count.reset_index(name='skills_count')

#to sort values according to skills_count column 
df_skills_count = df_skills_count.sort_values(by='skills_count', ascending=False)
df_skills_count
```

```python
#making individual vairables so that if we want to change it later it will make it easier
job_title = 'Data Analyst'
top_skills = 10

df_skill_final = df_skills_count[df_skills_count['job_title_short'] == job_title].head(top_skills)
df_skill_final
```

```python
#plotting
df_skill_final.plot(kind='barh', x='job_skills', y='skills_count')
plt.gca().invert_yaxis()
plt.title(f'Top {top_skills} Skills for {job_title}s')
plt.ylabel('')
plt.xlabel('Job Posting Count')
plt.legend().set_visible(False)
plt.show()
```

Average Salary by Skill

```python
df_exploded.head()
```

```python
average_salary_skill = df_exploded.groupby('job_skills')['salary_year_avg'].mean()


top_salary_skills = average_salary_skill.reset_index().sort_values(by='salary_year_avg', ascending=False).head(10)
top_salary_skills
```

```python
top_salary_skills.plot(kind='bar', x='job_skills', y='salary_year_avg')
plt.title('Average Salary By Skills for all Data Science Jobs')
plt.legend().set_visible(False)
plt.xticks(rotation=45, ha='right')
plt.show()
```
