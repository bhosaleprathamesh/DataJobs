```python
#Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup 
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.index.name = 'job_index'
```

```python
df_india = df[df['job_country'] == 'India']
df_india
```

```python
df_india.reset_index(inplace=True)
df_india
```

```python
df_india.set_index('job_index', inplace=True)
df_india
```

```python
median_pivot = df_india.pivot_table(index='job_title_short', values='salary_year_avg', aggfunc=['median', 'min', 'max'])
median_pivot
```

```python
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False)
```
