```python
# Importing Libraries
import ast
import seaborn as sns
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_ind = df[df['job_country'] == 'India']
df_ind
```

```python
df_na = df_ind.dropna(subset=['salary_year_avg'])
df_na
```

```python
job_titles = df_na['job_title_short'].value_counts().index[:3].tolist()

job_titles = sorted(job_titles)

job_titles
```

```python
fig, ax = plt.subplots(len(job_titles), 1)

for i, job_title in enumerate(job_titles):
    print(i, job_title)
```

```python
fig, ax = plt.subplots(len(job_titles), 1, figsize=(10, 10))

for i, job_title in enumerate(job_titles):
    df_plot = df_na[df_na['job_title_short'] == job_title]
    df_plot['salary_year_avg'].plot(kind='hist', ax=ax[i], bins=20, edgecolor='black', color='red', alpha=0.7)
    #ax[i].set_xlabel('Salary ($USD)')
    ax[i].set_title(f'Salary Distribution for {job_title}', fontsize=14)
    ax[i].set_ylabel('Number of Jobs')
    ax[i].set_xlim(0, 230000)
    ax[i].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
    ax[i].grid(True, linestyle='--', alpha=0.6)
    
    

plt.tight_layout()
plt.show()
```

```python
df_da = df_na[df_na['job_title_short'] == 'Data Analyst']
df_da
```

```python
df_da = df_na[df_na['job_title_short'] == 'Data Analyst']

sns.displot(df_da['salary_year_avg'], kind='kde', fill=True, color='red')
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f'${int(y/1000)}K'))

plt.title("Distribution of India's Data Analyst Yearly Salaries in $(USD)")
plt.xlabel('Yearly Salary')
plt.xlim(0, 250000)
plt.ylabel('Number of Jobs')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
