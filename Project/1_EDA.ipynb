### Exploratory Data Analysis

```python
# Importing Libraries
import ast
import seaborn as sns
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df
```

```python
df_plot = df['job_country'].value_counts().head(20).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_country', hue='count', palette='flare', legend=False)
plt.title('Number of Jobs per Country')
plt.ylabel('')
plt.xlabel('Number of Jobs')
sns.despine()
plt.show()
```

```python
df_plot = df['job_title_short'].value_counts().to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_title_short', hue='count', palette='flare', legend=False)
plt.title('Number of Jobs per Job Title for all Countries')
plt.ylabel('')
plt.xlabel('Number of Jobs')
sns.despine()
plt.show()
```

```python
df_plot = df['company_name'].value_counts().to_frame()[1:].head(20)

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='flare', legend=False)
plt.title('Number of Jobs per Company for all Countries')
plt.ylabel('')
plt.xlabel('Number of Jobs')
sns.despine()
plt.show()
```

### India

```python
df_india = df[df['job_country'] == 'India'].copy()
df_india
```

```python
df_plot = df_india['job_title_short'].value_counts().to_frame()

sns.barplot(data=df_plot, x='count', y='job_title_short', hue='count', palette='flare', legend=False)
plt.title('Data Jobs Count In India 2023')
plt.ylabel("")
plt.xlabel('Number of Jobs')
sns.despine()
plt.show()
```

### Data Analyst India

```python
df_da_india = df[(df['job_country'] == 'India') & (df['job_title_short'] == 'Data Analyst')]
df_da_india
```

```python
df_jobs = df_da_india['job_location'].value_counts().head(10).to_frame()

sns.set_style("ticks")
sns.barplot(data=df_jobs, x='count', y='job_location', hue='count', palette='flare', legend=False)
plt.title('Counts of Job Locations for Data Analyst in India')
plt.ylabel('')
plt.xlabel('Number of Jobs')
sns.despine()
plt.show()
```

```python
df_company = df_da_india['company_name'].value_counts().head(10).to_frame()

sns.set_style("ticks")
sns.barplot(data=df_company, x='count', y='company_name', hue='count', palette='flare', legend=False)
plt.title('Number of Jobs per Companies for Data Analyst in India')
plt.ylabel('')
plt.xlabel('Number of Jobs')
sns.despine()
```

```python
dict_column = {
    'job_work_from_home': 'Work From Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
}

for i, (column, title) in enumerate(dict_column.items()):
    print(i , column, title)
```

```python
fig, ax = plt.subplots(1, 2)
fig.set_size_inches((10, 5))
for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df_da_india[column].value_counts(), autopct='%1.1f%%', startangle=90, labels=['False', 'True'])
    ax[i].set_title(title)

fig.suptitle('For Data Analyst India', fontsize=17)
plt.show()
```
