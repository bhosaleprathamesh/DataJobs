```python
# Importing Libraries
import ast
import seaborn as sns
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_da_ind = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'India')].copy()
df_da_ind
```

```python
df_da_ind['job_month'] = df_da_ind['job_posted_date'].dt.month
df_da_ind
```

```python
df_da_ind_explode = df_da_ind.explode('job_skills')
df_da_ind_explode
```

```python
df_da_ind_explode.pivot_table(index='job_month', columns='job_skills', aggfunc='size', fill_value=0)
```

```python
df_da_ind_pivot = df_da_ind_explode.pivot_table(index='job_month', columns='job_skills', aggfunc='size', fill_value=0)
df_da_ind_pivot
```

```python
df_da_ind_pivot.loc['Total'] = df_da_ind_pivot.sum()
df_da_ind_pivot
```

```python
df_da_ind_pivot.loc['Total'].sort_values(ascending=False)
```

```python
df_da_ind_pivot = df_da_ind_pivot[df_da_ind_pivot.loc['Total'].sort_values(ascending=False).index]
df_da_ind_pivot
```

```python
df_da_ind_pivot = df_da_ind_pivot.drop('Total')
df_da_ind_pivot
```

```python
da_total = df_da_ind.groupby('job_month').size()
da_total
```

```python
df_da_ind_pivot.div(da_total/100, axis=0)
```

```python
df_da_ind_perc = df_da_ind_pivot.div(da_total/100, axis=0)
df_da_ind_perc
```

```python
df_da_ind_perc = df_da_ind_perc.reset_index()
df_da_ind_perc['job_posted_month'] = df_da_ind_perc['job_month'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_da_ind_perc = df_da_ind_perc.set_index('job_posted_month')
df_da_ind_perc = df_da_ind_perc.drop(columns='job_month')

df_da_ind_perc
```

```python
from matplotlib.ticker import PercentFormatter
df_plot = df_da_ind_perc.iloc[:, :5]

sns.lineplot(data=df_plot, dashes=False, palette='tab10', linewidth=2)
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills for Data Analysts in India', fontsize=15)
plt.ylabel('Likelihood of Job Posting')
plt.xlabel('2023')
plt.legend().remove()
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i])

plt.show()
```
